```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = [pd.read_parquet("../../data/highorder-data_1.parquet"),
 pd.read_parquet("../../data/allorders-data_1.parquet"),
 pd.read_parquet("../../data/pymhdfs-data_1.parquet")]
data[0]
```

```python
read = lambda key : 1e9*np.vstack([np.vstack(x[key]) for x in data])
```

```python
wfe_rms = 1e9*np.vstack([np.vstack(x["WfeRms"]) for x in data])
fig,ax = plt.subplots()
ax.plot(read("WfeRms"))
ax.grid()
ax.set_xlabel("Sample [#]")
```

```python
fig,ax = plt.subplots()
ax.plot(read("SegmentWfeRms"))
ax.grid()
ax.set_xlabel("Sample [#]")
```

```python
segment_wfe_rms = read("SegmentPiston")
fig,ax = plt.subplots()
ax.plot(segment_wfe_rms)
ax.grid()
ax.set_xlabel("Sample [#]")
ax.set_yticks(np.arange(-5,0)*550);
```

```python
fig,ax = plt.subplots()
ax.plot(segment_wfe_rms - segment_wfe_rms[:,-1][:,None])
ax.grid()
ax.set_xlabel("Sample [#]")
ax.set_yticks(np.arange(-5,0)*550);
```

```python
fig,ax = plt.subplots()
ax.plot(segment_wfe_rms.mean(1))
ax.grid()
ax.set_xlabel("Sample [#]")
```
